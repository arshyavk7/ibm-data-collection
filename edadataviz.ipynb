### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df and it contains the columns 'Orbit' and 'Class'
# 'Class' is assumed to be 1 for success and 0 for failure

# Step 1: Group by Orbit and calculate mean success rate
success_rate_by_orbit = df.groupby('Orbit')['Class'].mean().reset_index()

# Step 2: Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(data=success_rate_by_orbit, x='Orbit', y='Class', palette='viridis')
plt.title('Success Rate by Orbit Type')
plt.ylabel('Success Rate (Mean of Class)')
plt.xlabel('Orbit Type')
plt.ylim(0, 1)
plt.show()
```

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract year from the Date column (assuming it's string or datetime)
df['Year'] = df['Date'].str[:4]  # If Date is a string like '2010-06-04'
# If Date is datetime type, use: df['Year'] = df['Date'].dt.year

# Group by Year and calculate average success rate (Class)
yearly_success = df.groupby('Year')['Class'].mean().reset_index()

# Convert Year to int for proper plotting
yearly_success['Year'] = yearly_success['Year'].astype(int)

# Plotting
plt.figure(figsize=(10,6))
sns.lineplot(data=yearly_success, x='Year', y='Class', marker='o')
plt.title('Yearly Launch Success Rate Trend')
plt.xlabel('Year')
plt.ylabel('Average Success Rate')
plt.ylim(0, 1)  # Since success rate is between 0 and 1
plt.grid(True)
plt.show()
```

```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```
